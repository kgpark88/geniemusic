### 라이브러리 임포트

```python
import numpy as np
import pandas as pd
from pprint import pprint
from google_play_scraper import Sort, reviews_all
from app_store_scraper import AppStore
```

## 리뷰 데이터 수집(구글 플레이 스토어)

###  리뷰 데이터 수집 함수(구글 플레이스토어)

```python
def google_app_review_scrap(app):
    reviews = reviews_all(app, sleep_milliseconds=1, lang='ko', country='kr', sort=Sort.NEWEST, filter_score_with=None)
    df = pd.DataFrame(reviews)
    df['at'] =  pd.to_datetime(df['at'], format='%Y-%m-%d %H:%M:%S')
    df['source'] = 'google'
    file_name = app + '.csv'
    df.to_csv(file_name, index=False)
```

```python
# 앱이름 : com.ktmusic.geniemusic, com.iloen.melon, com.google.android.apps.youtube.music
google_app_review_scrap('com.ktmusic.geniemusic')
```

### 리뷰 데이터 전처리/분석

# 리뷰 데이터 수집(애플 앱 스토어)

```python
data = AppStore(country="kr", app_name="geniemusic")
data.review()
# data.review(how_many=20000)
```

```python
pprint(data.reviews)
```

```python
pprint(data.reviews_count)
```

```python
df = pd.DataFrame(np.array(data.reviews),columns=['review'])
df = df.join(pd.DataFrame(df.pop('review').tolist()))
df['date'] =  pd.to_datetime(df['date'], format='%Y-%m-%d %H:%M:%S')
df['source'] ='apple'
```

```python
df.isnull().sum()
```

```python
cols = ['isEdited', 'developerResponse', 'title']
df.drop(cols, axis=1, inplace=True)
```

```python
df.rename(columns = {'date':'at','review': 'content', 'rating':'score'},inplace=True)
```

```python
df.head()
```

```python
df.to_csv('apple_geniemusic_review.csv', index=False)
```
